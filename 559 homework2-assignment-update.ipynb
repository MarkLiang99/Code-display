Please display the category_id,the difference between the average film replacement cost and the average rental rate where the difference is larger than 17

```python
 cur.execute('''
 select film_category.category_id, AVG(film.replacement_cost)-AVG(film.rental_rate) AS difference 
from film
inner join film_category on film_category.film_id=film.film_id 
GROUP by film_category.category_id
Having difference >17

 ''').fetchall()
 
```

### <center> Question 7 (10 points)
Display the most frequently rented movies(film_id,title and count of rental times as times_rented) in descending order.

```python
 cur.execute('''
    SELECT film.film_id, film.title,count(film.title) as times_rented
From film
inner join inventory as i
on film.film_id = i.film_id
inner join rental
on i.inventory_id =rental.inventory_id
GROUP BY film.title
order BY times_rented DESC
 ''').fetchall()
```

### <center> Question 8 (10 points)
Many DVD stores produce a daily list of overdue rentals so that customers can be contacted and asked
to return their overdue DVDs.To create such a list, search the rental table for films with a return date that is NULL and where the rental date is further in the past than the rental duration specified in the film table. If so, the film is overdue and we should produce the name of the film along with the customer name and phone number.
(please use datetime function to add the rental_duration and compare the date with current date)

```python
 cur.execute('''
    SELECT film.title, customer.first_name,customer.last_name,address.phone
FROM film 
join inventory 
On film.film_id=inventory.film_id
join rental
on inventory.inventory_id=rental.inventory_id
join customer 
on rental.customer_id = customer.customer_id
join address 
on customer.address_id=address.address_id
Where rental.return_date is NULL or film.rental_duration < julianday(rental.return_date)-julianday(rental.rental_date)
 ''').fetchall()
```

### <center> Question 9 (10 points)
List the top five genres in gross revenue in descending order,please display category_name and revenue . (Hint: you may need to use the following tables: category, film_category, inventory, payment, and rental.)

```python
 cur.execute('''
    SELECT category.name,sum(payment.amount) as gross_revenue
From film
join inventory as i
on film.film_id = i.film_id
join rental
on i.inventory_id =rental.inventory_id
join payment 
on rental.rental_id=payment.rental_id
join film_category
on film_category.film_id = film.film_id
join category
on category.category_id= film_category.category_id
GROUP BY category.name
order BY gross_revenue DESC 
limit 5
 ''').fetchall()
```

### <center> Question 10 (20 points)
Write a query to display for each store its store ID, city, country and the total payment amount. The output must be sorted on store id.

```python
 cur.execute('''
    SELECT store.store_id,city.city,country.country,sum(payment.amount)
from store
join address
on store.address_id = address.address_id
join city 
on address.city_id= city.city_id
join country
on city.country_id=country.country_id
join payment
on store.manager_staff_id=payment.staff_id
group by payment.staff_id

 ''').fetchall()
```
