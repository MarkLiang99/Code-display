# Problem 4 (25 points total)

For this problem, you are to use the tables found in files “Roster.csv”, “ISE535 Spring 2021.csv”, “ISE599 Fall 2020.csv”, and “ISE535 Fall 2021.csv”

Perform the following steps:

a. Read the four csv files into dataframes and combine them into a single table that is in tidy format (call it “gradebook”). Convert the “Date of Birth” attribute to be a date type and covert the grade to a factor type. (15 points)


```python
#
# Insert Python code here
# 
Roster=read_csv('Roster.csv',skiprows=1)
ISE535_Spring=read_csv('ISE535 Spring 2021.csv',skiprows=2)
ISE599=read_csv('ISE599 Fall 2020.csv',skiprows=2)
ISE535=read_csv('ISE535 Fall 2021.csv',skiprows=2)
```

```python
gradebook=pd.merge(Roster,ISE535_Spring, how='outer',on=['Student ID'])
gradebook.rename(columns={'Grade':'ISE535_Spring_21'},inplace= True)
gradebook=pd.merge(gradebook,ISE599,how='outer', on=['Student ID'])
gradebook.rename(columns={'Grade':'ISE599_fall_20'},inplace= True)
gradebook=pd.merge(gradebook,ISE535,how='outer', on=['Student ID'])
gradebook.rename(columns={'Grade':'ISE535_fall_21'},inplace= True)
gradebook
```

```python
gradebook=gradebook.melt(id_vars=['Student ID','Student Name','Home Country','Date of Birth'],var_name='Crouse',value_name='Grade')
gradebook
```

```python
gradebook['Date of Birth']=pd.to_datetime(gradebook['Date of Birth'])
gradebook['Grade']=gradebook['Grade'].astype('category')
gradebook.dropna(inplace=True)
gradebook
```

b. Add a column that contains the grade point value for the class (A = 4, B = 3, C = 2) and then use that column to display the GPA for each student. Hint: look at the Tidyverse case_when command. (10 points)

```python
#
# Insert Python code here
# 
def convert (q):
    if q=='A':
        return 4
    if q=='B':
        return 3
    if q=='C':
        return 2
```

```python
qwe=list()
for i in gradebook['Grade']:
    qwe.append(convert(i))
qwe=np.array(qwe)
gradebook['GPA']=qwe
gradebook.groupby('Student Name')['GPA'].mean()
```
